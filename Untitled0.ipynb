```python
# Calculating daily caloric needs based on the Mifflin-St Joer Equation



def calculate_caloric_needs():

  gender = input("Enter your gender (male/female): "). strip().lower()
  weight = float(input("Enter your weight in kg: "))
  height = float(input("Enter your height in cm: "))
  age = int(input("Enter your age: "))

  print("Activity levels: ")
  print("1: Sedentary (little or no exercise)")
  print("2: Lightly active (light exercise/sports 1-3 days a week)")
  print("3: Moderately active (moderate exercise/sports 3-5 days a week)")
  print("4: Very active (hard exercise/sports 6-7 days a week)")
  print("5: Extra active (very hard exervise, physical job, or training twice a day)")

  activity_level = int(input("Choose your activity level (1-5): "))

  if gender == 'male':
    bmr = 10 * weight + 6.25 * height - 5 * age + 5
  else:
    bmr = 10 * weight + 6.25 * height - 5 * age - 161

  activity_factors = {
      1: 1.2,
      2: 1.375,
      3: 1.55,
      4: 1.725,
      5: 1.9
  }

  caloric_needs = bmr * activity_factors.get(activity_level, 1.2)

  return caloric_needs, weight, height, gender, age



def calculate_rdi(caloric_needs):
    # Macronutrient distribution ranges (percentages)
    carb_ratio = (0.45, 0.65)
    protein_ratio = (0.10, 0.35)
    fat_ratio = (0.20, 0.35)

    # 1 gram of carbohydrates = 4 kcal
    # 1 gram of proteins = 4 kcal
    # 1 gram of fats = 9 kcal

    # Calculate recommended intake for each macronutrient in grams
    carbs_min = (caloric_needs * carb_ratio[0]) / 4
    carbs_max = (caloric_needs * carb_ratio[1]) / 4
    proteins_min = (caloric_needs * protein_ratio[0]) / 4
    proteins_max = (caloric_needs * protein_ratio[1]) / 4
    fats_min = (caloric_needs * fat_ratio[0]) / 9
    fats_max = (caloric_needs * fat_ratio[1]) / 9

    # Return recommended dietary intake for macronutrients
    return {
        "carbs_min": carbs_min, "carbs_max": carbs_max,
        "proteins_min": proteins_min, "proteins_max": proteins_max,
        "fats_min": fats_min, "fats_max": fats_max
    }



def calculate_micronutrients(gender, weight, age):

  if gender == 'male':
    vitamin_c = 90 + 0.02 * weight
    calcium = 1000 + 0.02 * age
    iron = 8 + 0.1 * (weight/70)
    vitamin_d = 600
    magnesium = 400
    potassium = 3400
    vitamin_b12 = 2.4

  else:
    vitamin_c = 75 + 0.02 * weight
    calcium = 1000 + 0.02 * age
    iron = 18 + 0.1 * (weight/70)
    vitamin_d = 600
    magnesium = 310
    potassium = 2600
    vitamin_b12 = 2.4

  return {
        "Vitamin C (mg)": vitamin_c,
        "Calcium (mg)": calcium,
        "Iron (mg)": iron,
        "Vitamin D (IU)": vitamin_d,
        "Magnesium (mg)": magnesium,
        "Potassium (mg)": potassium,
        "Vitamin B12 (mcg)": vitamin_b12
    }

daily_calories, weight, height, sex, age = calculate_caloric_needs()
rdi = calculate_rdi(daily_calories)
micronutrients = calculate_micronutrients(sex, weight, age)

print(f"\nYour recommended daily caloric intake is: {daily_calories:.2f} kcal")
print(f"Recommended dietary intake (in grams per day):")
print(f"Carbohydrates: {rdi['carbs_min']:.2f}g - {rdi['carbs_max']:.2f}g")
print(f"Proteins: {rdi['proteins_min']:.2f}g - {rdi['proteins_max']:.2f}g")
print(f"Fats: {rdi['fats_min']:.2f}g - {rdi['fats_max']:.2f}g")

print("\nPersonalized micronutrient intake (per day):")
for nutrient, amount in micronutrients.items():
    print(f"{nutrient}: {amount:.2f}")


```
